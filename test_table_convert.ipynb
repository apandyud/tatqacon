```python
def detect_numbers(row):
    res = ''
    for item in row:
        if item.trim()  = ''
            res.append('S')
            continue
        if item.trim()  = '-'
            res.append('U')
            continue
            
        r = table_convert.extract_number(item)
        if r == None:
            res.append('O')
        else:
            res.append('N')
    return res           
```

```python
def extract_data_general(tables):
    results = []

    for table in tables:
        headers = table[0]
        context_stack = []

        for row in table[1:]:
            # Use the first cell of the row as a label if it's not empty
            label = row[0].strip()
            if label:
                context_stack = [label]

            # Iterate through the row's values (skipping the first label cell)
            for i in range(1, len(row)):
                value = row[i].strip()
                if value and value != '':
                    # Clean and convert the value
                    numeric_value = float(value.replace('$', '').replace(',', '').replace('(', '-').replace(')', ''))

                    # Gather contextual information
                    info = context_stack + [headers[i]]

                    results.append({
                        "value": numeric_value,
                        "info": [item for item in info if item]  # Exclude empty strings
                    })

    return results
```
