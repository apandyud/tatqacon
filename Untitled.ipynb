```python
devdf = pd.read_json('dataset_raw/tatqa_dataset_dev.json')
```

```python
res = []
for i, row in devdf.iterrows():
    for q in row['questions']:        
        if q['answer_type'] == 'arithmetic' and q['answer_from'] == 'table':            
            uid = row['table']['uid']
            #if uid not in res:
            #    res[uid] = {'table': row['table'], 'questions': []}
            #res[uid]['questions'].append(q)
            q['tuid'] = uid
            res.append(q)
qs_df = pd.DataFrame(res)
qs_df
```
